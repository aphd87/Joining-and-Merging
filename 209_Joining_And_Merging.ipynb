```python
#Merging

#Based on sharing a column
df_a = pd.DataFrame({"user_id": [1,2,3,5,7], "age" : [41,15,60,18,29]}) 
df_b = pd.DataFrame({"user_id": [1,2,3,4,5], "gender" : ["m","f","f","f","m"]})
```

```python
#Inner join

#return rows where there are values in both data frames

pd.merge(df_a,df_b, how = "inner", on = "user_id")
#As you can see, 1,2,3,5 are returned based on user_id as these observations
#All observations are located in both user_id columns
```

```python
#LEFT Join
#Use all rows from first/left data frame, we append on values from second data frame
#Finds overlap across both
pd.merge(df_a,df_b, how = "left", on = "user_id")
#All user ids from df_a are found here, as this is the first listed data frame
#gender is appened from df_b, where there was a user_id match
#Where there wasn't a match in which user_id from df_a was not found in df_b, there is a NaN
```

```python
#Right join is same but logic is in reverse
#Replace left with right in the line of code
```

```python
#Outer Join
#All rows included in output, otherwise left as NAN
#Return all rows when there is a match in either of the tables

pd.merge(df_a,df_b, how = "outer", on = "user_id")

#Left join and a right join
#All user_ids from both tables
#Any information that can be appended, is appended, or it is a NaN
```

```python
#Rename column
df_b.rename(columns = {"user_id":"customer_id"}, inplace = True)
```

```python
#Inner join with different column names

#Now, the first data frame is organized by user_id, and the second is organized by customer_id
pd.merge(df_a,df_b, how = "inner", left_on = "user_id", right_on = "customer_id")
```
